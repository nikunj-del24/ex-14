```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15]
}

df = pd.DataFrame(data)
print(df)

      Product  Price  Units_Sold
0      Laptop  55000          25
1      Mobile  20000          60
2      Tablet  30000          35
3  Headphones   2000          80
4  Smartwatch   8000          50
5      Camera  45000          15
```

```python
#Data Binning
bins1 = [0,10000,30000,60000]
labels1 = ['Low','Medium','High']

df['Price_Category'] = pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)

      Product  Price  Units_Sold Price_Category
0      Laptop  55000          25           High
1      Mobile  20000          60         Medium
2      Tablet  30000          35         Medium
3  Headphones   2000          80            Low
4  Smartwatch   8000          50            Low
5      Camera  45000          15           High
```

```python
#Binning Units Sold
bins2 = [0,30,60,100]
labels2 = ['Low','Medium','High']

df['Units_Sold_Category'] = pd.cut(df['Units_Sold'],bins=bins2,labels=labels2)
print(df)

      Product  Price  Units_Sold Price_Category Units_Sold_Category
0      Laptop  55000          25           High                 Low
1      Mobile  20000          60         Medium              Medium
2      Tablet  30000          35         Medium              Medium
3  Headphones   2000          80            Low                High
4  Smartwatch   8000          50            Low              Medium
5      Camera  45000          15           High                 Low
```

```python
#Data Formatting
#Convert Data Type
print(df.dtypes)
df['Units_Sold'] = df['Units_Sold'].astype(float)
print(df.dtypes)

Product                  object
Price                     int64
Units_Sold                int64
Price_Category         category
Units_Sold_Category    category
dtype: object
Product                  object
Price                     int64
Units_Sold              float64
Price_Category         category
Units_Sold_Category    category
dtype: object
```

```python
#Convert Text to Uppercase
df['Product'] = df['Product'].str.upper()
df

Product	Price	Units_Sold	Price_Category	Units_Sold_Category
0	LAPTOP	55000	25.0	High	Low
1	MOBILE	20000	60.0	Medium	Medium
2	TABLET	30000	35.0	Medium	Medium
3	HEADPHONES	2000	80.0	Low	High
4	SMARTWATCH	8000	50.0	Low	Medium
5	CAMERA	45000	15.0	High	Low
```
